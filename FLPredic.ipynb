# Predicting for folder

```python
def load_new_images(folder_path):
    images = []
    file_names = []


    # Iterate through all files in the folder
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        # Load the image using PIL
        image = Image.open(file_path)
        # Resize the image to match the input size of the model (32x32)
        image = image.resize((32, 32))
        # Convert the image to a NumPy array and normalize pixel values
        image = np.array(image) / 255.0
        # Append the image and file name to the lists
        images.append(image)
        file_names.append(file_name)

    # Convert the list of images and file names to NumPy arrays
    images = np.array(images)
    file_names = np.array(file_names)
    return images, file_names

# Replace 'folder_path' with the path to the folder containing your new images
folder_path = "/content/drive/MyDrive/Colab Notebooks/Flower Classification_1/flowers 2/test"
new_images, file_names = load_new_images(folder_path)

# Make predictions using the loaded model
predictions = loaded_model.predict(new_images)

# Get the predicted class for each image
predicted_classes = np.argmax(predictions, axis=1)

# Get the class names
class_names = ['daisy', 'dandelion', 'rose', 'sunflower', 'tulip']

# Print the predicted class for each image
for i, predicted_class in enumerate(predicted_classes):
    print(f"Image: {file_names[i]}, Predicted Class: {class_names[predicted_class]}")
```
